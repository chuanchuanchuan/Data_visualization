```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import xlrd
#Pokemon dataset from kaggle
#sa means special_attack sd means special defend name means Pokemons' name
sa=[]
sd=[]
name=[]
worksheet = xlrd.open_workbook('pokemon_alopez247.xlsx')
Data_sheet = worksheet.sheets()[0]
row = Data_sheet.nrows
for i in range(1,row):
    sa.append(Data_sheet.cell_value(i,8))
    sd.append(Data_sheet.cell_value(i,9))
    name.append(Data_sheet.cell_value(i,1))
sa = np.array(sa)
sd = np.array(sd)
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02
nullfmt = NullFormatter()         # no labels
rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom_h, width, 0.2]
rect_histy = [left_h, bottom, 0.2, height]

# start with a rectangular Figure
plt.figure(1, figsize=(8, 8))

axScatter = plt.axes(rect_scatter)
axHistx = plt.axes(rect_histx)
axHisty = plt.axes(rect_histy)
#plt axes确定范围
# no labels
axHistx.xaxis.set_major_formatter(nullfmt)
axHisty.yaxis.set_major_formatter(nullfmt)

# the scatter plot:
axScatter.scatter(sa, sd)

# now determine nice limits by hand:
binwidth = 0.25
xymax = max(np.max(np.abs(sa)), np.max(np.abs(sd)))
lim = (int(xymax/binwidth) + 1) * binwidth

axScatter.set_xlim((0, lim))
axScatter.set_ylim((0, lim))
axScatter.set_xlabel("Special_Attack")
axScatter.set_ylabel("Special_Defend")
axHistx.set_title("Pokemon Special_Attack and Special_Defend before Gen6")
bins = np.arange(0, lim + binwidth, binwidth)
axHistx.hist(sa, bins=bins)
axHisty.hist(sd,bins=bins, orientation='horizontal')
#hist直方图与柱状图的区别 bins计数 柱状给位置给高度
axHistx.set_xlim(axScatter.get_xlim())
axHisty.set_ylim(axScatter.get_ylim())
#find the king of sa and sd
sd_index=np.where(sd==np.max(sd))
sa_index=np.where(sa==np.max(sa))
print(sd_index,sa_index)
print(name[211],name[148])
#outstand the best sa and sd
axScatter.scatter(sa[211],sd[211],color='red')
axScatter.scatter(sa[148],sd[148],color='red')
axScatter.annotate(name[211],(sa[211],sd[211]))
axScatter.annotate(name[148],(sa[148],sd[148]))
from pylab import *
#创建子图 显示图片
axes([0.6,0.3,.08,.08])
xticks([]), yticks([])
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
im = mpimg.imread('Mewtwo.jpg')
imshow(im)
axes([0.12,0.64,0.08,0.08])
xticks([]), yticks([])
im = mpimg.imread('Shuckle.png')
imshow(im)
plt.show()
```
