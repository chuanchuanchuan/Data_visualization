```python
import  matplotlib.pyplot as plt
import numpy as np
import xlrd
plt.figure(figsize=(20,10), dpi=80)
plt.figure(1)
x_value = np.array(range(1997,2017,1))
x_pos = np.arange(len(x_value))
#data
GDP = []
worksheet = xlrd.open_workbook('GDP-fromworldbank.xlsx')
Data_sheet = worksheet.sheets()[0]
#choose 5 countries
row = [10,24,37,49,60]
name = ["United Arab Emirates","Bulgaria","Canada","Colombia","Denmark"]
cols = Data_sheet.ncols
for i in row:
    temp = []
    for j in range(cols-21,cols-1):
        temp.append(Data_sheet.cell_value(i,j)/1000000)
    GDP.append(temp)
GDP=np.array(GDP)
ax1 = plt.subplot(421)
plt.bar(x_pos,GDP[0],alpha=0.6)
plt.title("the GDP values and changes of GDP of"+name[0],fontsize=8)
ax1.set_ylabel("Million $")
ax1.set_xticks(x_pos)
ax1.set_xticklabels(x_value,fontsize=8)
ax1.plot(x_pos,GDP[0],color='red')
ax2 = plt.subplot(422)
plt.bar(x_pos,GDP[1],alpha=0.6)
plt.title("the GDP values and changes of GDP of"+name[1],fontsize=8)
ax2.set_ylabel("Million $")
ax2.set_xticks(x_pos)
ax2.set_xticklabels(x_value,fontsize=8)
ax2.plot(x_pos,GDP[1],color='red')
ax3 = plt.subplot(423)
plt.bar(x_pos,GDP[2],alpha=0.6)
plt.title("the GDP values and changes of GDP of"+name[2],fontsize=8)
ax3.set_ylabel("Million $")
ax3.set_xticks(x_pos)
ax3.set_xticklabels(x_value,fontsize=8)
ax3.plot(x_pos,GDP[2],color='red')
ax4 = plt.subplot(424)
plt.bar(x_pos,GDP[3],alpha=0.6)
plt.title("the GDP values and changes of GDP of"+name[3],fontsize=8)
ax4.set_ylabel("Million $")
ax4.set_xticks(x_pos)
ax4.set_xticklabels(x_value,fontsize=8)
ax4.plot(x_pos,GDP[3],color='red')
ax5 = plt.subplot(425)
plt.bar(x_pos,GDP[4],alpha=0.6)
plt.title("the GDP values and changes of GDP of"+name[4],fontsize=8)
ax5.set_ylabel("Million $")
ax5.set_xticks(x_pos)
ax5.set_xticklabels(x_value,fontsize=8)
ax5.plot(x_pos,GDP[4],color='red')
plt.show()
#
plt.figure(2)
f, (ax, ax2) = plt.subplots(2, 1, sharex=True)
city_pos = np.arange(5)
avg_gdp = [np.mean(GDP[i]) for i in range(5)]
# plot the same data on both axes
ax.bar(city_pos,avg_gdp,alpha=0.6)
ax.set_title("The average GDP of different countries")
ax.set_xticks(city_pos)
ax.set_xticklabels(name)
plt.text(1-0.25,avg_gdp[1],str(avg_gdp[1]//1),color='black')
plt.text(3-0.25,avg_gdp[3],str(avg_gdp[3]//1),color='black')
ax2.bar(city_pos,avg_gdp,alpha=0.6)
ax2.set_ylabel("Million $")
ax2.set_xlabel("Country")
ax2.set_xticks(city_pos)
ax2.set_xticklabels(name,rotation = 45)
# zoom-in / limit the view to different portions of the data
ax.set_ylim(max(avg_gdp)-100000,max(avg_gdp))  # outliers only
ax2.set_ylim(0,400000)  # most of the data
plt.text(2-0.25,avg_gdp[1],str(avg_gdp[2]//1),color='black')
# hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop='off')  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

d = .015 
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal


plt.show()
```
