```python
#percentage of lines spoken per season
#data from Internet
import numpy as np
import matplotlib.pyplot as plt
N = 10
Chandler = np.array([17,16.8,15.8,18.6,17.6,18.6,15.6,12.8,17.3,15.5])
Joey=np.array([13.5,15.5,14.7,16.4,17.7,16.4,17.3,17.5,16.4,16.6])
Monica=np.array([18.4,16.6,16.2,15.4,16.7,16.4,16.7,15.5,17.1,15.5])
Phoebe=np.array([13.4,14.2,15.4,15,15.1,14.3,15,14.7,14.7,15.7])
Rachel=np.array([17.8,16.1,17.5,18.3,16.8,17.9,20.2,20.6,17.5,17.2])
Ross=np.array([19.9,20.8,20.5,16.3,16.2,16.4,15.2,18.8,17.1,19.4])
x = np.arange(N)
x = x + 1
fig, ax = plt.subplots()
width=0.8
plt.ylim((0,110))
p1 = plt.bar(x,Chandler,width,hatch='/',color='lightpink')
p2 = plt.bar(x, Joey, width,
             bottom=Chandler,hatch='*',color='mediumvioletred')
p3 = plt.bar(x, Monica, width,
             bottom=Joey+Chandler,hatch='+',color='plum')
p4 = plt.bar(x, Phoebe, width,
             bottom=Monica+Joey+Chandler,hatch='-',color='deepskyblue')
p5 = plt.bar(x, Rachel, width,
             bottom=Phoebe+Monica+Joey+Chandler,hatch='.',color='limegreen')
p6 = plt.bar(x, Ross, width,
             bottom=Rachel+Phoebe+Monica+Joey+Chandler,hatch='|',color='tomato')
ax.set_xticks(x)
ax.set_ylabel("Lines%")
ax.set_xlabel("Seasons")
plt.title("percentage of lines spoken per season")
plt.legend(('Chandler', 'Joey', 'Monica','Phoebe','Rachel','Ross'),
           bbox_to_anchor=(1.05, 0.5), loc=3, borderaxespad=0)
#add annotation
for x1,x2,x3,x4,x5,x6,x_pos in zip(Chandler,Joey, Monica,Phoebe,Rachel,Ross,x):
    plt.text(x_pos-0.4,x1/2,str(x1)+"%",color='white')
    plt.text(x_pos-0.4,x1+x2/2,str(x2)+"%",color='white')
    plt.text(x_pos-0.4,x2+x1+x3/2,str(x3)+"%",color='white')
    plt.text(x_pos-0.4,x3+x2+x1+x4/2,str(x4)+"%",color='white')
    plt.text(x_pos-0.4,x4+x3+x2+x1+x5/2,str(x5)+"%",color='white')
    plt.text(x_pos-0.4,x5+x4+x3+x2+x1+x6/2,str(x6)+"%",color='white')
plt.show()
```
