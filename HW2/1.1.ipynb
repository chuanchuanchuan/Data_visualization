```python
import matplotlib.pyplot as plt
import numpy as np
#2_1_1 索福瑞收视率数据 2018电视台年平均收视率
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
#解决中文乱码
TV = ['北京卫视','湖南卫视','东方卫视','浙江卫视', \
      '江苏卫视','金鹰卡通','山东卫视','安徽卫视', \
      '卡酷少儿','黑龙江卫视','天津卫视','重庆卫视']
audience_rating = [0.222,0.214,0.210,0.208,0.187,\
                   0.124,0.117,0.091,0.091,0.079,\
                   0.078,0.059]
#data
y_pos = np.arange(len(TV))
fig,ax = plt.subplots()
ax.barh(y_pos,audience_rating,alpha=0.4,color='green')
ax.set_xlim(0,max(audience_rating)+0.1)
ax.set_yticks(y_pos)
ax.set_yticklabels(TV)
ax.set_xlabel('audience_rating')
ax.set_title('Average audience_rating for 2018')
#显示数据
for x,y in zip(audience_rating,y_pos):
    plt.text(x+0.012,y,'%.3f' % x,ha='center',va='bottom')


plt.show()
```
