```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
#显示中文
from pylab import mpl
from matplotlib.patches import Polygon
mpl.rcParams['font.sans-serif'] = ['KaiTi']
mpl.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(16,7))
#定义图的大小
#实例化Basemap对象，指定投影方式
#conformal: merc
map = Basemap(projection='merc',llcrnrlat=-5,urcrnrlat=55,\
             llcrnrlon=65,urcrnrlon=147,lon_0=104,lat_0=23)
#创建一个只显示中国地区的地图
#projection:表示规定的投影方式
#llcrnrlat 所需地图域左下角的纬度
#urcrnrlat 所需地图域右上角的纬度
#llcrnrlon 所需地图域左下角的经度
#urcrnrlon 所需地图域左下角的经度
#lon_0 lat_0 中心

map.drawcoastlines(linewidth=1)#海岸线
map.drawcountries(linewidth=1)#国家
map.fillcontinents(color='coral',lake_color='aqua')#给地图上色 大陆和海洋
map.drawmapboundary(fill_color='aqua')
parallels = np.arange(-90,90,10.)
map.drawparallels(parallels,labels=[1,0,0,0],fontsize=10)
#绘制纬线
meridians = np.arange(-180.,181.,20.)
map.drawmeridians(meridians,labels=[0,0,0,1],fontsize=10)
#绘制经线
map.readshapefile('gadm36_CHN_shp/gadm36_CHN_1','states',drawbounds=True)
map.readshapefile('gadm36_TWN_shp/gadm36_TWN_1','taiwan',drawbounds=True)
#给中国大陆加上鲜艳的颜色
ax = plt.gca()
for nshape,seg in enumerate(map.states):
    poly = Polygon(seg,facecolor='r')
    ax.add_patch(poly)
for nshape,seg in enumerate(map.taiwan):
    poly = Polygon(seg,facecolor='r')
    ax.add_patch(poly)
plt.show()
```
