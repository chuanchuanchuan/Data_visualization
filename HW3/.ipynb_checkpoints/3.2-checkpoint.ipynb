```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
#
# 
def get_marker_color(depth):
    if depth < 300.0:
        return ('hotpink')
    elif depth < 500.0:
        return ('deeppink')
    else:
        return ('mediumvioletred')

#csv
posi = pd.read_csv("quakes.csv") 
#
lats = np.array(posi['lat']) 
#
lons = np.array(posi['long']) 
#
mags = np.array(posi['mag'],dtype=float)
#
depths = np.array(posi['depth'],dtype=float)

#lower left
minlon,minlat = min(lons),min(lats)
#upper right
maxlon,maxlat = max(lons),max(lats)

fig = plt.figure(figsize=(14,10))
plt.title("earthquakes distribution")
#
map = Basemap(projection='merc', resolution = 'l', area_thresh = 1000.0,
              lat_0=0, lon_0=120,
              llcrnrlon=minlon-5, llcrnrlat=minlat-8,
              urcrnrlon=maxlon+10, urcrnrlat=maxlat+3)
map.drawcoastlines() #
map.drawcountries() #
map.fillcontinents(color = 'coral',lake_color='aqua')
map.drawmapboundary()
map.drawmeridians(np.arange(0, 360, 30))#
map.drawparallels(np.arange(-90, 90, 30))#
min_marker_size = 5
for lon, lat, mag,depth in zip(lons,lats,mags,depths):
    x,y = map(lon, lat) #
    msize = mag*min_marker_size #
    marker_string = get_marker_color(depth) #
    map.scatter(x, y,msize,color=marker_string)
plt.show()
```
